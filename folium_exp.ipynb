```python
import folium
```

```python
my_map = folium.Map(location = [18.1006, 78.0469],zoom_start = 4)

```

```python
my_map
```

```python
my_map.add_child(folium.LatLngPopup())
```

```python
main_marker = {'loc1':[30.717010, 76.876679],'label1':'chandigarh_quarter',
              'loc2':[32.194219, 76.351668],'label2':'Dharamsala_Nany_home'}
tooltips = "click"
folium.Marker(location = main_marker['loc1'],popup =main_marker['label1'],tooltip = tooltips,icon = folium.Icon(icon = 'angry',prefix = 'fa')).add_to(my_map)
folium.Marker(location = main_marker['loc2'],popup =main_marker['label2'],tooltip = tooltips).add_to(my_map)

my_map
    
```
